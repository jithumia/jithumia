```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("GAMMA.txt", sep=" ")
print(df)
```

```python
df.index += 1 
```

```python
df.index
```

```python
df.keys().unique()
```

```python
df_kbp_ad=df.loc[:,['KBP_AD']]
df_clinical_ad=df.loc[:,['CLINICAL_AD']]

df_kbp_rd=df.loc[:,['KBP_RD']]
df_clinical_rd=df.loc[:,['CLINICAL_RD']]

df_kbp_mu=df.loc[:,['KBP_MU']]
df_clinical_mu=df.loc[:,['CLINICAL_MU']]

df_kbp_complex=df.loc[:,['KBP_COMPLEX']]
df_clinical_complex=df.loc[:,['CLINICAL_COMPLEX']]

```

```python
#Gamma_passing_rate Relative 3% 3mm
ax=df_clinical_rd.plot(figsize=(15,10),markevery=25,color='darkorange',linewidth=2.0)
df_kbp_rd.plot(ax=ax,markevery=25,color='darkviolet',linestyle='dashed',linewidth=2.0)

plt.title('Gamma Analysis', fontsize=15)
plt.ylabel('Passing rate [3% 3mm]')
plt.xlabel('Plans')
plt.grid(True)
plt.xticks(range(0,14))
plt.yticks(np.arange(95, 100.5, step=0.5))
axes = plt.gca()
axes.set_xlim([0,14])
axes.set_ylim([97,100])
plt.legend()
plt.savefig('GAMMA_RD.pdf')  
```

```python
#MU
ax=df_clinical_mu.plot(figsize=(15,10),markevery=25,color='navy',linewidth=2.0)
df_kbp_mu.plot(ax=ax,markevery=25,color='r',linestyle='dashed',linewidth=2.0)

plt.title('Monitor Units Comparison', fontsize=15)
plt.ylabel('Monitor Units')
plt.xlabel('Plans')
plt.grid(True)
plt.xticks(range(0,14))
plt.yticks(np.arange(1200,2800, step=100))
axes = plt.gca()
axes.set_xlim([0,14])
axes.set_ylim([1200,2800])
plt.legend()
plt.savefig('MU.pdf')  
```

```python
#Gamma_passing_rate Absolute 3% 3mm
ax=df_clinical_ad.plot(figsize=(15,10),markevery=25,color='darkorange',linewidth=2.0)
df_kbp_ad.plot(ax=ax,markevery=25,color='darkviolet',linestyle='dashed',linewidth=2.0)

plt.title('Gamma Analysis', fontsize=15)
plt.ylabel('Passing rate [3% 3mm]')
plt.xlabel('Plans')
plt.grid(True)
plt.xticks(range(0,14))
plt.yticks(np.arange(75, 100.5))
axes = plt.gca()
axes.set_xlim([0,14])
axes.set_ylim([80,100])
plt.legend()
plt.savefig('GAMMA_AD.pdf')  
```

```python
#Complexity
ax=df_clinical_complex.plot(figsize=(15,10),markevery=25,color='magenta',linewidth=2.0,label="Clinical")
df_kbp_complex.plot(ax=ax,markevery=25,color='chocolate',linestyle='dashed',linewidth=2.0,label="KBP")

plt.title('Plan Complexity Analysis', fontsize=15)
plt.ylabel('Edge Metric')
plt.xlabel('Plans')
plt.grid(True)
plt.xticks(range(0,14))
plt.yticks(np.arange(0, 0.2001, step =0.01))
axes = plt.gca()
axes.set_xlim([0,14])
axes.set_ylim([0,0.2001])
plt.legend()
plt.savefig('complexity.pdf')  
```
